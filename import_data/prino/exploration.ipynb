```python
import pandas as pd
```

```python
df = pd.read_csv("prino_to_hitchmap.csv")
```

```python
df["datetime"] = df["datetime"].astype("datetime64[ns]")

plot = df['datetime'].groupby([df["datetime"].dt.year, df["datetime"].dt.month]).count().plot(kind="bar", title="Points per month", figsize=(50, 5))
plot.figure.savefig("points_per_month.png", dpi=300, bbox_inches="tight")
```
