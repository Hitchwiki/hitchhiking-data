```python
import sqlite3
import pandas as pd
import wget
import os
```

```python
url = 'https://hitchmap.com/dump.sqlite'
filename = 'dump.sqlite'
if os.path.exists(filename):
        os.remove(filename)
filename = wget.download(url)
```

# Points

```python
points = pd.read_sql('select * from points', sqlite3.connect(filename))
```

```python
points
```

```python
points.datetime = pd.to_datetime(points.datetime)
points["ride_datetime"] = points.datetime
```

```python
def get_displayed_time(row):
    """Use the actual time of the ride if available, otherwise use the time of the review."""
    if not pd.isnull(row.ride_datetime):
        return f'ðŸ•’ {row.ride_datetime.strftime(", %B %Y")}'
    elif not pd.isnull(row.datetime):
        return row.datetime.strftime(", %B %Y")
    else:
        return ""
```
