```python
import numpy as np
import folium
import xyzservices.providers as xyz
```

```python
Z = np.loadtxt(
    "/home/till/projects/hitchhiking/hitchmap-data/visualization/predictors/intermediate/map_TransformedTargetRegressorWithUncertainty_world_10.txt",
    dtype=float,
)
Z.shape
```

```python
Z
```

```python
lon, lat = 12, 48.875
zoom_start = 9


tiles = xyz.CartoDB.Positron
m = folium.Map(
    tiles=folium.TileLayer(no_wrap=True, tiles=tiles),
    attr="Dummy",
    zoom_start=zoom_start,
    location=[lat, lon],
)
```

```python
image = Z

folium.raster_layers.ImageOverlay(
    image=image,
    bounds=[[-56, -180], [80, 180]],
    colormap=lambda x: (1, 0, 0, x),
).add_to(m)
```
